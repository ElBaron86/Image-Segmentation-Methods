### Getting image file_names

```python
import matplotlib.pyplot as plt
import cv2

# Directory containing images
directory = './images'
files_jpg = [f for f in os.listdir(directory) if f.endswith('.jpg')]

plt.figure(figsize=(10, 3*len(files_jpg)))
for i, img_name in enumerate(files_jpg):

# Display the original and segmented images on subplots

    plt.subplot(len(files_jpg), 2, (i+1)*2 -1)
    plt.imshow(cv2.cvtColor(cv2.imread(f"images/{img_name}"), cv2.COLOR_BGR2RGB))
    plt.tight_layout()
    plt.title('Original Image')
    plt.axis('off') 

    plt.subplot(len(files_jpg), 2, (i+1)*2)
    plt.imshow(region_growing(cv2.imread(f"images/{img_name}")), cmap='gray')
    plt.tight_layout()
    plt.title('Segmented Image')
    plt.axis('off')

plt.show()
```
